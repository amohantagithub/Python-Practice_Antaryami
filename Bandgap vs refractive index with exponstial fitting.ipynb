```python
Eg = [0.1, 0.5, 1, 1.5, 2, 2.5, 3, 3.5, 4, 4.5, 5, 5.5, 6]
import numpy as np
np_Eg=np.array(Eg)
print(np_Eg)
```

```python
def func(x):
    y = (1 + (13.6/(x+3.4))**2)**0.5
    return y
z=func(np_Eg)
    
```

```python
import matplotlib.pyplot as plt
plt.scatter(np_Eg, z, color ='r')
plt.xlabel('Bandgap(eV)')
plt.ylabel('Refractive index')
plt.margins(0.02)
plt.show()
```

```python
def func(u,a,b,c):
    return a*np.exp(-b*u)+c
```

```python
from scipy.optimize import curve_fit
popt, pcov = curve_fit (func,np_Eg, z)
```

```python
popt
```

```python
plt.plot(np_Eg, func(np_Eg, *popt), 'r-', label='fit: a=%5.3f, b=%5.3f, c=%5.3f' % tuple(popt))
plt.show()
```

```python
plt.scatter(np_Eg, z, color ='r', label = 'data')
plt.plot(np_Eg, func(np_Eg, *popt), 'r-', label='fit: a=%5.3f, b=%5.3f, c=%5.3f' % tuple(popt))
plt.xlabel('Bandgap(eV)')
plt.ylabel('Refractive index')
plt.legend()
plt.margins(0.02)
plt.show()
```
